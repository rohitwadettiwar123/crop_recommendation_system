# 🌾 Crop Recommendation — Exploratory Data Analysis
**Explainable Crop Suitability Prediction System**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
import warnings
warnings.filterwarnings('ignore')
plt.style.use('dark_background')
print('Libraries loaded')
```

## 1. Load Dataset

```python
df = pd.read_csv('../data/Crop_recommendation.csv')
print(f'Shape: {df.shape}')
df.head()
```

## 2. Basic Info & Missing Values

```python
print(df.dtypes)
print('\nNull Values:\n', df.isnull().sum())
print('\nDuplicates:', df.duplicated().sum())
```

## 3. Descriptive Statistics

```python
feat_cols=['N','P','K','temperature','humidity','ph','rainfall']
stats=df[feat_cols].describe().T
stats['skewness']=df[feat_cols].skew()
stats['kurtosis']=df[feat_cols].kurt()
stats.round(3)
```

## 4. Crop Distribution

```python
cc=df['label'].value_counts()
fig=px.bar(x=cc.index,y=cc.values,title='Crop Distribution',color=cc.values,color_continuous_scale='Greens')
fig.show()
```

## 5. Feature Histograms

```python
fig,axes=plt.subplots(2,4,figsize=(18,8))
for i,col in enumerate(feat_cols):
    axes[i//4][i%4].hist(df[col],bins=40,color='#2ecc71',alpha=0.8)
    axes[i//4][i%4].set_title(col)
plt.tight_layout(); plt.show()
```

## 6. Correlation Heatmap

```python
corr=df[feat_cols].corr()
fig=go.Figure(go.Heatmap(z=corr.values,x=corr.columns,y=corr.columns,colorscale='RdYlGn',zmid=0,text=corr.round(2).values,texttemplate='%{text}'))
fig.update_layout(title='Pearson Correlation',height=480)
fig.show()
```

## 7. Outlier Detection

```python
for col in feat_cols:
    Q1,Q3=df[col].quantile([0.25,0.75])
    IQR=Q3-Q1
    n=((df[col]<Q1-1.5*IQR)|(df[col]>Q3+1.5*IQR)).sum()
    print(f'{col:15s}: {n} outliers')
```

## 8. Per-Crop Mean Values

```python
df.groupby('label')[feat_cols].mean().round(2)
```
